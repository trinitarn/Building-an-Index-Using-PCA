```python
import pandas as pd

# Load the dataset
file_path = "b3a_pk2_4.csv"
data = pd.read_csv("b3a_pk2_4.csv")

# Function to split the pk18 values and count occurrences for each character
def count_pk18_combinations(row):
    # Get the pk18 value, which is a string
    pk18_value = row['pk18']
    
    # Split the pk18 string into individual characters and return each character
    return list(pk18_value)

# Apply the function to split pk18 and create a new DataFrame where each row represents one of the individual characters
expanded_data = data.explode('pk18').reset_index(drop=True)

# Now group by 'hhid07', 'pk2type', and 'pk18' and count the occurrences
frequency_table = expanded_data.groupby(['hhid07', 'pk2type', 'pk18']).size().reset_index(name='frequency')

# Pivot the table to create a cross-tabulation of pk2type (rows) and pk18 (columns)
pivot_table = frequency_table.pivot_table(index='pk2type', columns='pk18', values='frequency', aggfunc='sum', fill_value=0)

# Filter to only show A, B, C, D in the columns (you can adjust this as needed)
pivot_table = pivot_table[['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'O', 'P']]  # Add or modify categories here
```

```python
import pandas as pd
df1 = pd.read_csv("b3a_pk2_4.csv", dtype={"hhid07": str})

df2 = df1[["hhid07", "pk18", "pk2type"]]
```

```python
df2.head()
```

```python
# Define a function to apply the condition
def value_condition(pk18):
    if 'B' in pk18:
         return 1
    else:
        return 0

df2['value'] = df2['pk18'].apply(value_condition)

print(df2)
```

```python
import pandas as pd

df3 = df2[["hhid07", "pk2type", "value"]]
```

```python
# First, group by 'hhid07', and then pivot the data
df_grouped = df2.groupby(['hhid07', 'pk2type'])['value'].first().reset_index()

# Now, pivot the data from long to wide
df_wide = df_grouped.pivot(index=["hhid07"], columns="pk2type", values="value")

# If you want to reset the column names (make them clean)
df_wide.columns = df_wide.columns.to_series().str.strip()

# Reset index to make it flat if desired (optional)
df_wide.reset_index(inplace=True)

# Display the reshaped DataFrame
print(df_wide)
```
